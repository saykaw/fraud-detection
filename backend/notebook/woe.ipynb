```python
df = pd.read_csv('../data/fraudTrain.csv')
```

```python
target_col = 'is_fraud'

# Function to calculate WOE
def calculate_woe(group):
    total_count = len(group)
    positive_count = sum(group[target_col])
    negative_count = total_count - positive_count
    if positive_count == 0 or negative_count == 0:
        return 0  # Return zero if one of the counts is zero
    return np.log((positive_count / total_count) / (negative_count / total_count))

# Calculate WOE for category
category_woe = {}
category_grouped = df.groupby('category')
for category, group in category_grouped:
    category_woe[category] = calculate_woe(group)

# Calculate WOE for job
job_woe = {}
job_grouped = df.groupby('job')
for job, group in job_grouped:
    job_woe[job] = calculate_woe(group)

# Calculate WOE for city
city_woe = {}
city_grouped = df.groupby('city')
for city, group in city_grouped:
    city_woe[city] = calculate_woe(group)

# Print or store the WOE values for each category, job, and city
print("WOE for each category:")
print(category_woe)

print("\nWOE for each job:")
print(job_woe)

print("\nWOE for each city:")
print(city_woe)
```
